## 1. 기본 배열 만들기

치트시트는 1차원, 2차원, 3차원 배열 예제로 시작합니다. NumPy의 핵심은 **동일한 자료형을 가진 다차원 배열**을 빠르게 다룰 수 있다는 점입니다.  
아래에서는 `a`를 1차원 배열, `b`를 2차원 배열, `c`를 3차원 배열로 만들어 보겠습니다.

```python
# 1차원, 2차원, 3차원 NumPy 배열을 생성하는 예제입니다.
a = np.array([1, 2, 3])
b = np.array([(1.5, 2, 3), (4, 5, 6)], dtype=float)
c = np.array([[(1.5, 2, 3), (4, 5, 6)],
              [(3, 2, 1), (4, 5, 6)]], dtype=float)

print('a =\n', a)
print('b =\n', b)
print('c =\n', c)
```

### 실행 결과 예시

- `a`는 `[1 2 3]` 형태의 1차원 배열로 출력됩니다.  
- `b`는 2행 3열 구조의 2차원 배열로 보입니다.  
- `c`는 2개의 2차원 배열을 포함한 3차원 배열로 출력됩니다.

이 단계에서 중요한 점은, NumPy 배열은 차원 수가 올라가도 **하나의 일관된 객체**로 다뤄진다는 것입니다.

## 2. 초기화용 배열 만들기

실무에서는 데이터를 바로 입력하기보다, 먼저 원하는 모양의 배열을 만든 뒤 값을 채우는 경우가 많습니다.  
치트시트에는 0으로 채운 배열, 1로 채운 배열, 일정 간격 배열, 균등 분할 배열, 상수 배열, 단위행렬, 난수 배열, 빈 배열이 소개됩니다.

```python
# 0, 1, 등차수열, 균등분할, 상수, 단위행렬, 난수, 빈 배열을 만드는 예제입니다.
zeros_arr = np.zeros((3, 4))
ones_arr = np.ones((2, 3, 4), dtype=np.int16)
print('np.zeros((3,4)) =\n', zeros_arr)
print('np.ones((2,3,4), dtype=np.int16) =\n', ones_arr)
```

```python
d = np.arange(10, 25, 5)
lin = np.linspace(0, 2, 9) # step: (2-0)/8 = .25

print('np.arange(10,25,5) =', d)
print('np.linspace(0,2,9) =', lin)
```

## 4. 배열 수학 연산

NumPy의 가장 강력한 장점 중 하나는 **반복문 없이 배열 전체에 연산을 적용**할 수 있다는 점입니다.  
치트시트에는 덧셈, 뺄셈, 곱셈, 나눗셈뿐 아니라 지수, 제곱근, 삼각함수, 로그, 내적도 포함되어 있습니다.

```python
# 배열 간 사칙연산과 지수, 제곱근, 삼각함수, 로그, 내적을 계산하는 예제입니다.
print(a)
print(b)
print('a - b =\n', a - b)
print('np.subtract(a, b) =\n', np.subtract(a, b))
print('b + a =\n', b + a)
print('np.add(b, a) =\n', np.add(b, a))
print('a / b =\n', a / b)
print('np.divide(a, b) =\n', np.divide(a, b))
print('a * b =\n', a * b)
print('np.multiply(a, b) =\n', np.multiply(a, b))
```

```python
print('np.exp(b) =\n', np.exp(b))
print('np.sqrt(b) =\n', np.sqrt(b))
print('np.sin(a) =', np.sin(a))
print('np.cos(b) =\n', np.cos(b))
print('np.log(a) =', np.log(a))
```

```python
print(f)
print('e.dot(f) =\n', e.dot(f))
```

### 실행 결과 예시

- `np.exp(b)`는 각 원소에 지수 함수를 적용합니다.

## 5. 집계 함수

집계 함수는 배열 전체나 특정 축(axis)을 기준으로 요약 통계를 계산할 때 사용합니다.  
합계, 최솟값, 최댓값, 누적합, 평균, 중앙값, 상관계수, 표준편차는 데이터 분석에서 매우 자주 등장합니다. [file:1]

```python
print(a)
print(b)
```

```python
# 합계, 최솟값, 최댓값, 누적합, 평균, 중앙값, 상관계수, 표준편차를 계산합니다.
print('a.sum() =', a.sum())
print('a.min() =', a.min())
print('b.max(axis=0) =', b.max(axis=0))
print('b.cumsum(axis=1) =\n', b.cumsum(axis=1))
print('a.mean() =', a.mean())
print('np.median(b) =', np.median(b))
print('np.corrcoef(a) =\n', np.corrcoef(a))
print('np.std(b) =', np.std(b))
```

### 실행 결과 예시

- `a.sum()`은 전체 합계인 6을 반환합니다.  
- `b.max(axis=0)`는 열(column) 기준 최댓값을 구합니다.  
- `b.cumsum(axis=1)`는 각 행에서 왼쪽부터 누적합을 계산합니다.  
- `np.median(b)`는 중앙값, `np.std(b)`는 표준편차를 구합니다.  
- `np.corrcoef(a)`는 자기 자신과의 상관을 계산하므로 1에 가까운 값이 나옵니다.

## 8. 정렬

치트시트에는 `a.sort()`와 `c.sort(axis=0)` 예제가 나옵니다.  
정렬은 축(axis)을 기준으로 동작하므로, 어떤 방향으로 정렬되는지 이해하는 것이 중요합니다.

```python
# 1차원, 2차원, 3차원 NumPy 배열을 생성하는 예제입니다.
a = np.array([3, 2, 1])
print('a =\n', a)
# 1차원 배열과 다차원 배열을 정렬하는 예제입니다.
a_sorted = a.copy()
a_sorted.sort()
print('a.sort() ->', a_sorted)
```

```python
c = np.array(np.arange(12, 0, -1)).reshape(2, 2, 3)
print('c =\n', c, "\n")
c_sorted = c.copy()
c_sorted.sort(axis=0) #  모든 원소에서 슬라이스끼리 정렬
c_sorted
```

```python
c = np.array(np.arange(12, 0, -1)).reshape(2, 2, 3)
print('c =\n', c, "\n")
c_sorted = c.copy()
c_sorted.sort(axis=1) # 슬라이스 내에서 행끼리 정렬
c_sorted
```

```python
c = np.array(np.arange(12, 0, -1)).reshape(2, 2, 3)
print('c =\n', c, "\n")
c_sorted = c.copy()
c_sorted.sort(axis=2) # 행 내에서 열끼리 정렬
c_sorted
```

### axis=0 — 슬라이스 방향 정렬
- 고정: [j, k] (행·열 위치) → 비교: c[0][j,k] vs c[1][j,k]
- 각 [j,k] 위치에서 슬라이스끼리 값을 비교해 오름차순 재배치합니다. 결과적으로 sorted[0]에 각 위치의 min이, sorted[1]에 max가 모입니다. 이 예제에서는 c[0]의 모든 값(7~12)이 c1보다 크므로 슬라이스 전체가 뒤집힌 것처럼 보이지만, 알고리즘은 어디까지나 element-wise입니다.

### axis=1 — 행 방향 정렬
- 고정: [i, k] (슬라이스·열 위치) → 비교: c[i][0,k] vs c[i][1,k]
- 각 슬라이스 내부에서, 열 위치를 고정하고 행들 사이 값을 비교합니다. 슬라이스 간 교환은 전혀 없고, 슬라이스 0과 슬라이스 1이 각각 독립적으로 정렬됩니다.

### axis=2 — 열 방향 정렬
- 고정: [i, j] (슬라이스·행 위치) → 비교: 해당 행의 원소들 c[i,j,0..2]
- 가장 직관적인 케이스입니다. 각 행을 독립적으로 오름차순 정렬하고, 슬라이스 구조와 행 구조는 그대로 유지됩니다. 2D 배열에서 axis=1로 정렬하는 것과 완전히 동일한 개념이 각 슬라이스에 반복 적용되는 것입니다.

### 실행 결과 예시

- 1차원 배열은 작은 값부터 큰 값 순으로 정렬됩니다.  
- 다차원 배열은 `axis=0`이면 첫 번째 축을 따라 각 위치별로 정렬됩니다.  
정렬은 데이터 전처리나 순위 계산 전에 자주 사용됩니다.

```python
c = np.array([[(1.5, 2, 3), (4, 5, 6)],
              [(3, 2, 1), (4, 5, 6)]], dtype=float)
print('c =\n', c,  "\n")

c_sorted = c.copy()
c_sorted.sort(axis=0)
print('c.sorted(axis=0) =\n', c_sorted)
```

## 9. 부분집합, 슬라이싱, 인덱싱

NumPy 배열도 파이썬 리스트처럼 인덱싱과 슬라이싱이 가능합니다.  
다만 2차원 이상에서는 행과 열을 동시에 선택할 수 있기 때문에 훨씬 강력합니다. [file:1]

```python
# 1차원, 2차원, 3차원 NumPy 배열을 생성하는 예제입니다.
a = np.array([1, 2, 3])
b = np.array([(1.5, 2, 3), (4, 5, 6)], dtype=float)
c = np.array([[(1.5, 2, 3), (4, 5, 6)],
              [(3, 2, 1), (4, 5, 6)]], dtype=float)

print('a =\n', a)
print('b =\n', b)
print('c =\n', c)
```

```python
# 특정 원소 선택, 구간 슬라이싱, 행/열 선택, 역순 조회를 보여 줍니다.
print('a[2] =', a[2])
print('b[1,2] =', b[1, 2], b[1][2])
print('a[0:2] =', a[0:2])
print('b[0:2,1] =', b[0:2, 1])
print('b[:1] =\n', b[:1])
print('c[1,...] =\n', c[1, ...])
print('a[::-1] =', a[::-1])
```

### 실행 결과 예시

- `a[2]`는 세 번째 원소인 3을 선택합니다.  
- `b[1,2]`는 2행 3열 위치의 값인 6.0을 선택합니다.  
- `a[0:2]`는 처음 두 원소를 가져옵니다.  
- `a[::-1]`는 배열을 역순으로 뒤집습니다.  
- `c[1,...]`에서 `...`는 나머지 모든 축을 의미합니다.

## 10. 불리언 인덱싱과 팬시 인덱싱

인덱싱은 NumPy의 핵심 활용 포인트입니다.  
조건으로 원소를 고르는 불리언 인덱싱과, 원하는 위치를 직접 지정하는 팬시 인덱싱을 익히면 데이터 추출이 훨씬 쉬워집니다.

```python
print('a =\n', a)
print('b =\n', b)
```

```python
# 조건에 맞는 원소를 선택하는 불리언 인덱싱과 위치 기반 팬시 인덱싱 예제입니다.
print('a[a < 2] =', a[a < 2])
print('b[[1,0,1,0],[0,1,2,0]] =', b[[1,0,1,0],[0,1,2,0]])
print('b[[1,0,1,0]][:,[0,1,2,0]] =\n', b[[1,0,1,0]][:,[0,1,2,0]])
```

```python
b[[1,0,1,0]]
```

```python
b[[1,0,1,0]][:,[0,1,2,0]]
```

### 실행 결과 예시

- `a[a < 2]`는 조건을 만족하는 값만 남겨 `[1]`이 됩니다.  
- `b[[1,0,1,0],[0,1,2,0]]`는 지정한 좌표의 값을 한 번에 가져옵니다.  
- 두 번째 예제는 먼저 행을 고른 뒤, 다시 열을 선택해 부분행렬을 만드는 방식입니다.

## 11. 배열 형태 변경

데이터를 분석하다 보면 같은 데이터라도 모양(shape)을 바꿔야 할 때가 많습니다.  
전치, 평탄화, reshape는 데이터를 재배열하는 대표적인 연산입니다.

```python
a, b
```

```python
# 전치(transpose), 평탄화(ravel), 모양 변경(reshape)을 수행합니다.
i = np.transpose(b)
print('np.transpose(b) =\n', i)
print('i.T =\n', i.T)
print('b.ravel() =', b.ravel())
```

```python
g = a - b
print(g)
print('g.reshape(3, -2) =\n', g.reshape(3, -2))
```

### 실행 결과 예시

- `np.transpose(b)`는 행과 열을 뒤바꿉니다.  
- `i.T`도 전치와 같은 의미입니다.  
- `b.ravel()`은 다차원 배열을 1차원으로 펼칩니다.  
- `reshape(3, -2)`는 한쪽 크기를 자동 계산해 새로운 shape으로 바꿉니다.  
실무에서는 머신러닝 입력 형태를 맞출 때 자주 사용합니다.

## 12. 요소 추가와 삭제

배열의 모양을 바꾸거나 원소를 추가, 삽입, 삭제하는 기능도 치트시트에 포함되어 있습니다.  
다만 NumPy 배열은 리스트처럼 가변 길이 구조가 아니므로, 내부적으로는 새 배열이 만들어지는 경우가 많습니다.

```python
g
```

```python
# 배열 크기 변경과 원소 추가/삽입/삭제를 수행하는 예제입니다.
h = np.array([1, 2, 3, 4, 5, 6])
h.resize((2, 3))
print('resized h =\n', h)
print('np.append(h, g) =', np.append(h, g))
```

```python
print(a)
print('np.insert(a, 1, 5) =', np.insert(a, 1, 5))
print('np.delete(a, [1]) =', np.delete(a, [1]))
```

```python
k = np.arange(10).reshape(2, 5)
print(k)
p = np.append(k, np.arange(5))
print(p)
```

### 실행 결과 예시

- `resize((2,3))`는 배열 shape을 2행 3열로 바꿉니다.  
- `np.append`는 배열을 1차원으로 바꾸어 추가합니다.  
- `np.insert(a, 1, 5)`는 인덱스 1 위치에 5를 삽입합니다.  
- `np.delete(a, [1])`는 두 번째 원소를 삭제한 새 배열을 만듭니다.

## 13. 배열 결합

여러 배열을 하나로 합치는 작업도 매우 중요합니다.  
행 방향 결합, 열 방향 결합, 단순 이어 붙이기 등은 데이터 병합의 기초입니다.

```python
print(a)
print(b)
print(d)
print(e)
print(f)
```

```python
# 여러 배열을 세로, 가로, 열 방향으로 결합하는 방법을 보여 줍니다.
print('np.concatenate((a, d), axis=0) =', np.concatenate((a, d), axis=0))
print('np.vstack((a, b)) =\n', np.vstack((a, b)))
print('np.r_[e, f] =\n', np.r_[e, f])
print('np.hstack((e, f)) =\n', np.hstack((e, f)))
print('np.column_stack((a, d)) =\n', np.column_stack((a, d)))
print('np.c_[a, d] =\n', np.c_[a, d])
```

### 실행 결과 예시

- `np.concatenate((a, d), axis=0)`는 1차원 배열 두 개를 이어 붙입니다.  
- `np.vstack((a, b))`는 위아래로 쌓아 2차원 배열을 만듭니다.  
- `np.hstack((e, f))`는 좌우로 붙입니다.  
- `np.column_stack((a, d))`와 `np.c_[a, d]`는 열 기준으로 묶어 2열짜리 배열을 만듭니다.

## 14. 배열 분할

배열 결합의 반대 작업이 분할입니다.  
큰 배열을 여러 조각으로 나누어 처리할 때 사용합니다.

```python
a
```

```python
c
```

```python
# 배열을 가로와 세로 방향으로 분할하는 예제입니다.
print('np.hsplit(a, 3) =\n', np.hsplit(a, 3))
print('np.vsplit(c, 2) =\n', np.vsplit(c, 2))
```

### 실행 결과 예시

- `np.hsplit(a, 3)`는 1차원 배열을 3개 조각으로 나눕니다.  
- `np.vsplit(c, 2)`는 첫 번째 축 기준으로 나누어 3차원 배열을 둘로 쪼갭니다.  
이런 기능은 데이터 배치(batch)를 나누거나, 훈련/검증 데이터를 분리할 때 응용할 수 있습니다.

## 17. 치트시트와 실제 사용법의 차이

치트시트는 빠르게 훑어보기 좋은 자료이지만, 실제 코드에서는 약간 다르게 쓰는 부분도 있습니다.  
예를 들어 치트시트에는 `b.median()` 또는 `a.corrcoef()`처럼 보이는 표기가 있지만, 일반적인 NumPy 사용에서는 `np.median(b)`, `np.corrcoef(a)`처럼 함수 형태가 더 자연스럽습니다.
