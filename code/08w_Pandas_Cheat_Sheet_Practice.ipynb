## 2. Reshaping Data – 레이아웃 변경, 정렬, 인덱싱, 이름 변경

```python
df = pd.DataFrame({"a": [4,5,6], "b": [7,8,9], "c": [10,11,12]}, index=[1,2,3])

# Melt: 컬럼 → 행
melted = pd.melt(df)
print(melted)
```

```python
# Pivot: 행 → 컬럼
pivoted = melted.pivot(columns="variable", values="value")
print(pivoted)
```

```python
df1 = pd.DataFrame({"a": [1, 2]})
df2 = pd.DataFrame({"a": [3, 4]})

# concat: 행 방향으로 붙이기
print(pd.concat([df1, df2]))
```

```python
df1 = pd.DataFrame({"a": [1, 2]})
df2 = pd.DataFrame({"b": [3, 4]})

# concat: 컬럼 방향으로 붙이기
print(pd.concat([df1, df2], axis=1))
```

## 3. Method Chaining – 메서드 연결

```python
df = pd.DataFrame({"a": [4,5,6], "b": [700,8,900], "c": [10,11,12]}, index=[1,2,3])
df
```

```python
result = (pd.melt(df)
    .rename(columns={"variable": "var", "value": "val"})
    .query("val >= 200"))
print(result)
```

```python
result = pd.melt(df) \
           .rename(columns={"variable": "var", "value": "val"}) \
           .query("val >= 200")
print(result)
```

## 9. 그룹 데이터 (GroupBy)

```python
df = pd.DataFrame({
    "col":   ["A","B","A","B","A"],
    "value": [10, 20, 30, 40, 50]})

grp = df.groupby(by="col")
print(grp.size())
print(grp.sum())
print(grp.mean())
print(grp.agg(["min","max","mean"]))
```

```python
# GroupBy 내 순위/이동 함수
df = pd.DataFrame({"col": ["A","A","B","B"], "val": [3,1,4,2]})
df["rank_dense"]  = df.groupby("col")["val"].rank(method="dense")
df["rank_min"]    = df.groupby("col")["val"].rank(method="min")
df["rank_pct"]    = df.groupby("col")["val"].rank(pct=True)
df["rank_first"]  = df.groupby("col")["val"].rank(method="first")
df["shift_1"]     = df.groupby("col")["val"].shift(1)
df["shift_m1"]    = df.groupby("col")["val"].shift(-1)
df["cumsum"]      = df.groupby("col")["val"].cumsum()
print(df)
```

## 10. 새 컬럼 생성

```python
df = pd.DataFrame({
    "Length": [3, 5, 7],
    "Height": [2, 4, 6],
    "Depth":  [1, 2, 3]})

df = df.assign(Area=lambda d: d.Length * d.Height)
df["Volume"] = df.Length * df.Height * df.Depth
df["Bucket"] = pd.qcut(df.Length, q=2, labels=False)
print(df)
```

## 11. 결측값 처리

```python
df = pd.DataFrame({
    "a": [1, None, 3],
    "b": [None, 5, None],
    "c": [7, 8, 9]})

print(df.dropna())            # NA 있는 행 삭제
print(df.fillna(0))           # NA를 0으로 채우기
print(pd.isnull(df))          # NA 여부 확인
print(pd.notnull(df))         # 비NA 여부 확인
```

## 12. 데이터 합치기 (Joins)

```python
adf = pd.DataFrame({"x1": ["A","B","C"], "x2": [1,2,3]})
bdf = pd.DataFrame({"x1": ["A","B","D"], "x3": ["T","F","T"]})

print("Left Join:")
print(pd.merge(adf, bdf, how="left",  on="x1"))

print("Right Join:")
print(pd.merge(adf, bdf, how="right", on="x1"))

print("Inner Join:")
print(pd.merge(adf, bdf, how="inner", on="x1"))

print("Outer Join:")
print(pd.merge(adf, bdf, how="outer", on="x1"))
```

```python
# Filtering Joins
print("adf에서 bdf에 있는 x1 행:")
print(adf[adf.x1.isin(bdf.x1)])

print("adf에서 bdf에 없는 x1 행:")
print(adf[~adf.x1.isin(bdf.x1)])
```

```python
# Set-like Operations
ydf = pd.DataFrame({"x1":["A","B","C"], "x2":[1,2,3]})
zdf = pd.DataFrame({"x1":["B","C","D"], "x2":[2,3,4]})

print("교집합(Intersection):")
print(pd.merge(ydf, zdf))

print("합집합(Union):")
print(pd.merge(ydf, zdf, how="outer"))

print("차집합(Setdiff - ydf only):")
print(
    pd.merge(ydf, zdf, how="outer", indicator=True)
    .query('_merge == "left_only"')
    .drop(columns=["_merge"])
)
```

## 13. 윈도우 함수 (Windows)

```python
df = pd.DataFrame({"val": [1, 2, 3, 4, 5, 6]})

df["expanding_sum"] = df["val"].expanding().sum()
df["rolling_mean3"] = df["val"].rolling(3).mean()
df["cumsum"]  = df["val"].cumsum()
df["cummax"]  = df["val"].cummax()
df["cummin"]  = df["val"].cummin()
df["cumprod"] = df["val"].cumprod()
print(df)
```

## 14. 타입 변환 (Changing Type)

```python
df = pd.DataFrame({
    "num_str":  ["1","2","3"],
    "date_str": ["2024-01-01","2024-06-15","2024-12-31"],
    "category": ["a","b","a"]})

df["num"]      = pd.to_numeric(df["num_str"])
df["date"]     = pd.to_datetime(df["date_str"])
df["category"] = df["category"].astype("category")
print(df.dtypes)
print(df)
```

## 15. Datetime 속성

```python
s = pd.to_datetime(pd.Series(["2024-03-15", "2025-11-30", "2026-07-04"]))
df = pd.DataFrame({"date": s})
df["year"]    = df["date"].dt.year
df["month"]   = df["date"].dt.month
df["day"]     = df["date"].dt.day
df["hour"]    = df["date"].dt.hour
df["minute"]  = df["date"].dt.minute
df["second"]  = df["date"].dt.second
df["quarter"] = df["date"].dt.quarter
print(df)
```

## 16. 문자열 연산 (Series.str)

```python
s = pd.Series(["Hello World", "pandas library", "Data Science"])

print(s.str.count("a"))           # 'a' 개수
print(s.str.len())                # 문자열 길이
print(s.str.upper())              # 대문자
print(s.str.lower())              # 소문자
print(s.str.title())              # 단어 첫글자 대문자
print(s.str.replace("a","@"))     # 치환
print(s.str.slice(0, 5))          # 슬라이싱
print(s.str.split(" ").str.get(0))  # 분리 후 첫 단어
print(s.str.join("-"))            # 비워두면 에러, 리스트 원소 연결용
print(s.str.isalnum())            # 영숫자 여부
```

## 17. Mapping / Apply

```python
s = pd.Series([1, 2, 3, 4, 5])
print(s.map(lambda x: x * 2))

df = pd.DataFrame({"a": [1,2,3], "b": [4,5,6]})
print(df.apply(lambda s: s.max() - s.min(), axis=1))
```

## 18. 시각화 (Plotting)

> matplotlib 백엔드를 사용합니다.

```python
import matplotlib
matplotlib.use("Agg")
import matplotlib.pyplot as plt

df = pd.DataFrame({
    "x": range(10),
    "y": [i**2 for i in range(10)],
    "z": [i*3 for i in range(10)]})

fig, axes = plt.subplots(2, 3, figsize=(14, 8))

df.plot(x="x", y="y", ax=axes[0,0], title="Line")
df.plot.scatter(x="x", y="y", ax=axes[0,1], title="Scatter")
df[["y","z"]].plot.hist(ax=axes[0,2], bins=5, title="Histogram", alpha=0.7)
df[["y","z"]].plot.bar(ax=axes[1,0], title="Bar")
df[["y","z"]].plot.area(ax=axes[1,1], title="Area", alpha=0.5)
df[["y","z"]].plot.box(ax=axes[1,2], title="Boxplot")

plt.tight_layout()
plt.savefig("pandas_plots.png", dpi=100)
plt.show()
print("저장 완료: pandas_plots.png")
```

## 19. 입출력 (Input/Output)
